### Método 1: Usando un ciclo for

```python
import numpy as np

# Usando la libreria numpy, ordenamos alfabeticamente los diferentes barrios
# y los guardamos en una variable

barrios = np.sort(inmuebles.Barrio.unique())
barrios
```

```python
# Creamos un dataframe vacio
area_estadistica = pd.DataFrame()

# Y recorremos cada barrio
for barrio in barrios:
  query = inmuebles.Barrio == barrio
  sub_dt = inmuebles[query]

  # Calculamos algunas cantidades estadísticas
  calculos = sub_dt[['Area']].describe().T
  calculos.index = [barrio]

  # Y concatenamos con el dataframe vacio anterior
  area_estadistica = pd.concat([area_estadistica, calculos], axis = 0)

area_estadistica.index.name = 'Barrio'
area_estadistica
```

###Método 2: Usando groupby

```python
area_estadistica = inmuebles[['Barrio','Area']].groupby('Barrio').describe()
area_estadistica.columns = [dato for _,dato in area_estadistica.columns]

area_estadistica
```

###Creando gráficos

Creamos una columna nueva con el precio del metro cuadrado en cada caso.

```python
inmuebles['Precio_por_m2'] = inmuebles.Precio_Millon / inmuebles.Area

inmuebles.sample(10)
```

Obtenemos algunos datos estadístico del precio por metro cuadrado $m^2$

```python
datos_est = inmuebles.groupby(by = 'Barrio').describe()

datos_est.sample(10)
```

```python
datos_est.Precio_por_m2.sample(10)
```

Tomaremos el top 10 de promedios, con su respectiva desviación estándar

```python
top_promedio_porM2 = datos_est.Precio_por_m2[['mean', 'std']].sort_values(by = 'mean', ascending=False).head(10)
top_promedio_porM2 = top_promedio_porM2.reset_index()
top_promedio_porM2
```

Vamos a probar varias graficas para encontrar la mejor

LinePlot

```python
sns.set_theme()
plt.figure(figsize = (10,8))

sns.lineplot(data = top_promedio_porM2, y = 'mean', x = 'Barrio', marker = 'o')

sigma = top_promedio_porM2['std']
x = top_promedio_porM2['Barrio']
y = top_promedio_porM2['mean']
plt.errorbar(x = x, y = y, yerr=(sigma, sigma), fmt='none', color='r', capsize=5)

plt.title('Promedio del costo por m2 por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio')
plt.xticks(rotation = 80)
plt.show()
```

Barplot

```python
# Reiniciamos settings
sns.set_theme()
plt.figure(figsize = (10,8))

sns.barplot(data = top_promedio_porM2, y = 'mean', x = 'Barrio')

#Obtenemos la desviación estandar
sigma = top_promedio_porM2['std']

x = top_promedio_porM2['Barrio']
y = top_promedio_porM2['mean']

# Graficamos las barras de error
plt.errorbar(x = x, y = y, yerr=(sigma, sigma), fmt='none', color='k', capsize=5)

plt.title('Top 10 Promedio del costo por m2 por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio (millones COP)')

# Rotamos las etiquetas de x para que los nombres no se solapen
plt.xticks(rotation = 80)

plt.savefig('/content/drive/MyDrive/Inmersion_datos_alura/img/top10_primedio_costo_m2.png', format = 'png')
plt.show()
```
